```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
# import iris
from sklearn.datasets import load_iris

raw = load_iris()

data = pd.DataFrame(raw.data, columns=raw.feature_names)
data.head()
```

```python
data['target'] = raw.target

# split data
X_train, X_test, y_train, y_test = train_test_split(data[raw.feature_names], data['target'], random_state=0)

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train, y_train)

knn.score(X_test, y_test)

```

```python
# size of the figure
plt.figure(figsize=(10,5))
# plot the data to subplot 1
plt.figure(1)
plt.subplot(1,2,1)
# real value of test data
plt.title('Real')
plt.scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=y_test, cmap='winter')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
# plt.show()
# plot redirects to a subplot 2
plt.subplot(1,2,2)
# predicted value of test data
plt.title('Predicted')
plt.scatter(X_test['sepal length (cm)'], X_test['sepal width (cm)'], c=knn.predict(X_test), cmap='winter')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```
